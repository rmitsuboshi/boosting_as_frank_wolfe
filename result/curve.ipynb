# Generates the figures shown in this paper.

```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['figure.subplot.left'] = 0.15
```

```python
BOOSTERS = [
    "lpb",
    "erlpb",
    "mlpb(ss)",
    "mlpb(pfw)",
]


DATASET = [
    "banana",
    "breast_cancer",
    "diabetis",
    "flare_solar",
    "german",
    "heart",
    "image",
    "ringnorm",
    "splice",
    "thyroid",
    "titanic",
    "twonorm",
    "waveform",
]

TITLES = {
    "banana":        "Banana",
    "breast_cancer": "B.Cancer",
    "diabetis":      "Diabetes",
    "flare_solar":   "F.Solar",
    "german":        "German",
    "heart":         "Heart",
    "image":         "Image",
    "ringnorm":      "Ringnorm",
    "splice":        "Splice",
    "thyroid":       "Thyroid",
    "titanic":       "Titanic",
    "twonorm":       "Twonorm",
    "waveform":      "Waveform",
}

ALGO_NAMES = {
    "mlpb(ss)":  "Short step",
    "mlpb(pfw)": "Pairwise",
    "lpb":       "LPB.",
    "erlpb":     "ERLPB.",
}


LINE_STYLES = {
    "mlpb(ss)":  "solid",
    "mlpb(pfw)": "solid",
    "lpb":       "solid",
    "erlpb":     "solid",
}

ALGO_NAMES = {
    "lpb":        "LPB.",
    "erlpb":      "ERLPB.",
    "mlpb(ss)":   "MLPB. (SS)",
    "mlpb(pfw)":  "MLPB. (PFW)",
    "mlpb(ss_only)":   "MLPB. (SS only)",
    "mlpb(pfw_only)":  "MLPB. (PFW only)",

}


LINE_COLORS = {
    "lpb":       "blue",
    "erlpb":     "green",
    "mlpb(ss)":  "red",
    "mlpb(pfw)": "purple",
    "mlpb(ss_only)":  "red",
    "mlpb(pfw_only)": "purple",
}
```

```python
for dataset in DATASET:
    plt.figure()
    df = dict()
    for i, booster in enumerate(BOOSTERS):
        file = f"./margin/{booster}_{dataset}.csv"
        df[booster] = pd.read_csv(file)
        df[booster]["time"] = df[booster]["time"].map(lambda t: t / 1000.0)



    for booster in BOOSTERS:

        x = df[booster]["time"]
        y = df[booster]["objval"]
        # plt.plot(x, y, label=ALGO_NAMES[booster], color=COLORS[booster], fillstyle='none')
        plt.plot(x, y, label=ALGO_NAMES[booster], color=LINE_COLORS[booster], fillstyle='none', linestyle=LINE_STYLES[booster])

    plt.legend(loc="lower right")
    plt.grid()
    plt.xscale('log')
    plt.ylim(bottom=-0.2)
    plt.title(TITLES[dataset], fontsize=16)
    plt.xlabel("Time", fontsize=16)
    plt.ylabel("Soft margin objective", fontsize=16)
    plt.savefig(f"../../figure/curve_logtime_{dataset}.pdf")
    # plt.savefig(f"margin/curve_logtime_{dataset}.pdf")
```

```python
lpb_count = list()
fw_iter = list()
plt.figure()
for dataset in DATASET:
    file = f"./margin/mlpb(pfw)_{dataset}.csv"
    df = pd.read_csv(file)
    lpb_count.append(df["lpb_call"].sum())
    fw_iter.append(df["lpb_call"].map(lambda x: 1 if x == 0 else 0).sum())

labels = [TITLES[name] for name in DATASET]
plt.bar(labels, fw_iter, label="FW update", edgecolor='black', fill=False, hatch='//')
plt.bar(labels, lpb_count, label="LPB update", bottom=fw_iter, edgecolor='black', fill=False, hatch='xxx')
plt.ylabel("Count")

plt.xticks(rotation=30)
plt.legend()
plt.savefig("margin/count.pdf")
plt.show()
```

```python
for dataset in DATASET:
    plt.figure()    

    n = 0
    for booster in ["mlpb(ss)", "mlpb(pfw)"]:
        file = f"./margin/{booster}_{dataset}.csv"
        df = pd.read_csv(file)
        x = [i for i in range(df.shape[0])]
        y = df["lpb_call"].cumsum()
        plt.plot(x, y, label=ALGO_NAMES[booster], fillstyle='none', linestyle=LINE_STYLES[booster], color=LINE_COLORS[booster])
        n = max(df.shape[0], n)

    t = np.arange(0.0, n, 0.1)
    plt.plot(t, t, color="black", linestyle="dotted")


    plt.grid()
    plt.legend()
    # plt.xscale('log')

    # plt.ylim()
    plt.title(TITLES[dataset], fontsize=18)
    plt.xlabel("Number of iterations", fontsize=18)
    plt.ylabel("Total LPB. update", fontsize=18)
    
    plt.savefig(f"margin/mlpb_lpb_update_{dataset}.pdf")

    # plt.savefig(f"margin/test_logtime_{dataset}.pdf")
```

### Compare `MLPBoost` with `ShortStep`, `ShortStep` only, `Pairwise`, and `Pairwise` only.

```python
for dataset in DATASET:
    plt.figure()

    max_n = 0
    for booster in ["mlpb(ss)", "mlpb(pfw)"]:
        file = f"./margin/{booster}_{dataset}.csv"
        df = pd.read_csv(file)

        n = df.shape[0]
        max_n = max(n, max_n)


        x = [i for i in range(n)]
        y = df["lpb_call"].cumsum()
        plt.plot(x, y, label=ALGO_NAMES[booster], color=LINE_COLORS[booster], fillstyle='none', linestyle=LINE_STYLES[booster])

    t = np.arange(0.0, n, 0.1)
    plt.plot(t, t, color="black", linestyle="dashed")

    plt.grid()
    # plt.xscale('log')

    # plt.ylim()
    plt.legend(loc="upper left")
    plt.title(TITLES[dataset], fontsize=18)
    plt.xlabel("iterations", fontsize=18)
    plt.ylabel("LPB. updates", fontsize=18)
    plt.savefig(f"../../arxiv/figure/mlpb_lpb_update_{dataset}.pdf")
    # plt.savefig(f"margin/mlpb_lpb_update_{dataset}.pdf")
```

```python
BOOSTERS = [
    "mlpb(ss)",
    "mlpb(ss_only)",
    "mlpb(pfw)",
    "mlpb(pfw_only)",
]
LINE_STYLES = {
    "mlpb(ss)": "solid",
    "mlpb(ss_only)": "dashed",
    "mlpb(pfw)": "solid",
    "mlpb(pfw_only)": "dashed",
}

for dataset in DATASET:
    plt.figure()
    ymax = -1.0
    df = dict()
    for i, booster in enumerate(BOOSTERS):
        file = f"./margin/{booster}_{dataset}.csv"
        df[booster] = pd.read_csv(file)
        df[booster]["time"] = df[booster]["time"].map(lambda t: t / 1000.0)
        ymax = max(ymax, df[booster]["objval"].max())

    ymax += 0.20
    for booster in BOOSTERS:
        x = df[booster]["time"]
        y = df[booster]["objval"]
        # plt.plot(x, y, label=ALGO_NAMES[booster], color=COLORS[booster], fillstyle='none')
        plt.plot(x, y, label=ALGO_NAMES[booster], color=LINE_COLORS[booster], fillstyle='none', linestyle=LINE_STYLES[booster])

    plt.legend(loc="upper center", ncol=2)
    plt.grid()
    plt.xscale('log')
    if dataset == "thyroid":
        plt.ylim(bottom=-0.4, top=ymax)
    else:
        plt.ylim(bottom=-0.2, top=ymax)
    plt.title(TITLES[dataset], fontsize=16)
    plt.xlabel("Time", fontsize=16)
    plt.ylabel("Soft margin objective", fontsize=16)
    # plt.savefig(f"margin/compare_fw_{dataset}.pdf")
    plt.savefig(f"../../figure/compare_fw_{dataset}.pdf")
```
